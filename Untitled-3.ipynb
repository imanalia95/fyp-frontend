```python
import pandas as pd

df = pd.read_csv("lecturers_clean_2.csv")
```

```python
collection = client.create_collection("lecturers")
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

documents = df['profile'].tolist()
embeddings = model.encode(documents, normalize_embeddings=True).tolist()

metadatas = []
ids = []

for i, row in df.iterrows():
    metadatas.append({
        "name": row["name"],
        "summary": row["summary"],
        "num_articles": row["num_articles"],
        "num_proceedings": row["num_proceedings"],
        "top_articles": row["top_articles"],
        "google_scholar": row["google_scholar"],
        "img_url": row["img_url"]   
    })
    ids.append(str(i))

collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=ids
)
```

```python
import chromadb
from config import CHROMA_PATH, embedding_fn

client = chromadb.PersistentClient(path=CHROMA_PATH)

collection = client.get_or_create_collection(
    name="lecturers",
    embedding_function=embedding_fn
)

data = collection.get(limit=1)

print(data["metadatas"][0].keys())
print(data["documents"][0][:1000])
```
