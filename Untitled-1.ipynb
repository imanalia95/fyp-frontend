```python
import pandas as pd
import json
import re
```

```python
def clean(text):
    """Remove messy whitespace and escape chars."""
    if pd.isna(text):
        return ""
    return re.sub(r'\s+', ' ', re.sub(r'\\r\\n', '; ', str(text))).strip()
```

```python
def build_profile(row):
    """
    Combine summary (weighted 2x) + article/proceeding titles.
    Summary is repeated to give it higher importance in embeddings.
    """
    parts = []

    # Summary x2 = higher semantic weight
    summary = clean(row['summary'])
    if summary:
        parts.append(summary)
        parts.append(summary)  # repeat for emphasis

    # Add publication titles (these contain domain keywords)
    for col in ['articles', 'proceedings']:
        pub_text = clean(row[col])
        if pub_text:
            # Take first 600 chars to avoid overly long profiles
            parts.append(pub_text[:600])

    return ' '.join(parts)
```

```python
df = pd.read_csv('all_lecturers.csv')
print(f"Loaded {len(df)} lecturers")
```

```python
df['profile'] = df.apply(build_profile, axis=1)

# Drop rows with empty profiles (lecturers with no data)
df_clean = df[df['profile'].str.strip() != ''].copy()
df_clean = df_clean.reset_index(drop=True)
print(f"After cleaning: {len(df_clean)} lecturers with profiles")
```

```python
df_clean.to_csv('lecturers_clean.csv', index=False)
print("Saved: lecturers_clean.csv")
```

```python
print("\n--- Sample profiles ---")
for _, row in df_clean.head(3).iterrows():
    print(f"\n👤 {row['name']}")
    print(f"   Profile (first 200 chars): {row['profile'][:200]}...")
```

```python
import pandas as pd
df = pd.read_csv('all_lecturers.csv')
columns_to_check = ['summary', 'articles', 'proceedings', 'books', 'chapters']
```

```python
def char_count(text):
    if pd.isna(text):
        return 0
    return len(str(text))
```

```python
def word_count(text):
    if pd.isna(text):
        return 0
    # Split by whitespace
    return len(str(text).split())
```

```python
for col in columns_to_check:
    df[f'{col}_chars'] = df[col].apply(char_count)
    df[f'{col}_words'] = df[col].apply(word_count)
```

```python
df['total_chars'] = df[[f'{col}_chars' for col in columns_to_check]].sum(axis=1)
df['total_words'] = df[[f'{col}_words' for col in columns_to_check]].sum(axis=1)
```

```python
print(df[[*columns_to_check, 
          *[f'{col}_chars' for col in columns_to_check], 
          *[f'{col}_words' for col in columns_to_check],
          'total_chars', 'total_words']].head())
```

```python
import pandas as pd

# -------- Load CSV --------
df = pd.read_csv("all_lecturers.csv")

# -------- Helper: count using ';' --------
def count_items(text):
    if pd.isna(text) or str(text).strip() == "":
        return 0
    
    return len([t for t in str(text).split(";") if t.strip() != ""])

# -------- Apply counts --------
df["num_articles"] = df["articles"].apply(count_items)
df["num_proceedings"] = df["proceedings"].apply(count_items)
df["num_books"] = df["books"].apply(count_items)
df["num_chapters"] = df["chapters"].apply(count_items)

# -------- Save back --------
df.to_csv("all_lecturers.csv", index=False)

print("Done! Counts added successfully.")
```
